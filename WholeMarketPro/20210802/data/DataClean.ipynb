 ## 1. 去除"指数"数据

```python
# 提取300000.SZ之前的数据
df_head = df[df['kdcode'] < '300001.SZ']
# 保存300000.SZ之后的数据
df_tail = df[df['kdcode'] >= '300001.SZ']
# 处理掉df_head中股票代码含有.SH的数据
df_head = df_head[df_head['kdcode'].str.contains('.SZ')]
# 连接
stockDf = pd.concat([df_head, df_tail], axis = 0)
```

```python
stocks_code = stockDf['kdcode'].unique()
stocks_num = len(stocks_code)
"股票数量",stocks_num
```

```python
# 筛选需要的列
cols = ['kdcode','dt','volume','turnover','turn','pct_chg','vwap','trade_status','adj_close','adj_open','adj_high','adj_low']
df = stockDf[cols]
df
```

## 2. 数据清洗

按照[当前股市](http://data.eastmoney.com/zjlx/list.html)中含有的股票筛出数据

```python
# 爬取http://data.eastmoney.com/zjlx/list.html中所有股票的股票代码
from urllib.parse import quote
import requests

url = 'http://push2.eastmoney.com/api/qt/clist/get?cb=jQuery1123038200043529008054_1627897140633&fid=f184&po=1&pz=4541&pn=1&np=1&fltt=2&invt=2&fields=f12&ut=b2884a393a59ad64002292a3e90d46a5&fs=m%3A0%2Bt%3A6%2Bf%3A%212%2Cm%3A0%2Bt%3A13%2Bf%3A%212%2Cm%3A0%2Bt%3A80%2Bf%3A%212%2Cm%3A1%2Bt%3A2%2Bf%3A%212%2Cm%3A1%2Bt%3A23%2Bf%3A%212%2Cm%3A0%2Bt%3A7%2Bf%3A%212%2Cm%3A1%2Bt%3A3%2Bf%3A%212'
data = requests.get(url)
res=data.text
res = res[118:-4]
List = eval(res)
codes = []

for i in range(len(List)):
    codes.append(List[i].get('f12'))

len(codes)
```

```python
#从所有数据中筛选codes中的股票
clean_df = pd.DataFrame(columns=df.columns)
for i, code_i in enumerate(codes):
    if i % 100 == 0:
        print(i)
    da = df[df['kdcode']==code_i+".SZ"]
    if da.empty:
        da = df[df['kdcode']==code_i+".SH"]
    if len(da) > 200:
        clean_df = pd.concat([da, clean_df])
```

```python
clean_df
```

```python
clean_df.set_index(['dt'], inplace = True)
clean_df.index = pd.to_datetime(clean_df.index)
# tmpdf = clean_df.copy()
```
